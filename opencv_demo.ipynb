1.使用 Tensorflow 打开一张图片

```python
%matplotlib inline
import matplotlib.pyplot as plt
import tensorflow as tf   
import numpy as np

image_raw_data = tf.gfile.FastGFile('italy_worldcup.jpg','r').read()
    
with tf.Session() as sess:
    img_data = tf.image.decode_jpeg(image_raw_data)
    plt.figure(dpi=150)
    plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
    plt.imshow(img_data.eval())
    plt.show()
```

2.使用 OpenCV 打开一张图片

```python
%matplotlib inline
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('italy_worldcup.jpg',cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) 
plt.figure(dpi=150)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

3.使用 OpenCV 处理一张图片

```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('italy_worldcup.jpg',0)
laplacian = cv2.Laplacian(img,cv2.CV_64F)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
plt.figure(dpi=150)
plt.subplot(2,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2),plt.imshow(laplacian,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()
```

4.使用 OpenCV 对图片中人脸进行检测

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
import cv2

font = cv2.FONT_HERSHEY_SIMPLEX
face_cascade = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
img = cv2.imread('italy_worldcup.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, 1.3, 5)
for index in range(faces.shape[0]):
    (x,y,w,h) = faces[index]
    # draw rectangle
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    # draw index 
    cv2.putText(img, str(index), (x+5,y-5), font, 0.75, (0,255,0), 2, cv2.LINE_AA)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = img[y:y+h, x:x+w]

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(dpi=150)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```
